# Init

```python
import pandas as pd
```

# Data Cleaning Quali

```python
jed24_q = pd.read_csv('saudi24_quali_laps.csv')
jed24_q
```

## Picking Data

```python
df1_jed24_q = jed24_q[['Driver', 'DriverNumber', 
                       'LapNumber','Stint', 'LapTime',
                       'Sector1Time', 'Sector2Time', 'Sector3Time',
                       'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST',
                       'Compound', 'TyreLife', 'FreshTyre',
                       'Deleted']]
df1_jed24_q            
```

## Convert to seconds

```python
# Convert string representations of time to Timedelta objects
df1_jed24_q['LapTime'] = pd.to_timedelta(df1_jed24_q['LapTime'])
df1_jed24_q['Sector1Time'] = pd.to_timedelta(df1_jed24_q['Sector1Time'])
df1_jed24_q['Sector2Time'] = pd.to_timedelta(df1_jed24_q['Sector2Time'])
df1_jed24_q['Sector3Time'] = pd.to_timedelta(df1_jed24_q['Sector3Time'])

# Now convert the Timedelta objects to seconds
df1_jed24_q['LapTime'] = df1_jed24_q['LapTime'].dt.total_seconds()
df1_jed24_q['Sector1Time'] = df1_jed24_q['Sector1Time'].dt.total_seconds()
df1_jed24_q['Sector2Time'] = df1_jed24_q['Sector2Time'].dt.total_seconds()
df1_jed24_q['Sector3Time'] = df1_jed24_q['Sector3Time'].dt.total_seconds()

df1_jed24_q
```

```python
df1_jed24_q.info()
```

## Cleaning

```python
df2_jed24_q = df1_jed24_q[df1_jed24_q['LapTime'].notna()]
df2_jed24_q
```

```python
df3_jed24_q = df2_jed24_q[df1_jed24_q['SpeedFL'].notna()]
df3_jed24_q
```

## Select every driver who went under 100s (actual outlap)

```python
df4_jed24_q = df3_jed24_q[df3_jed24_q['LapTime'] <= 90.0]
df4_jed24_q
```

```python
# Calculate the fastest lap time
fastest_lap_time = df3_jed24_q['LapTime'].min()

# Calculate 110% of the fastest lap time
one_ten_percent_fastest = 1.10 * fastest_lap_time

# Filter out every driver who did within 110% of the fastest lap time
df5_jed24_q = df3_jed24_q[df3_jed24_q['LapTime'] <= one_ten_percent_fastest]

df5_jed24_q
```

## Join

```python
df5_jed24_q
```

```python
quali = pd.read_csv('saudi24_quali_results.csv')
quali
```

```python
merge_df = pd.merge( df5_jed24_q, quali[['Abbreviation', 'TeamColor', 'TeamName']], left_on='Driver', right_on='Abbreviation', how='inner')
merge_df
```

```python
df6_jed24_q = merge_df 
df6_jed24_q
```

## Extract CSV for Readability

```python
df6_jed24_q.to_csv('jed24_q_clean.csv', index=False)
```

# Data Race Cleaning

```python
jed24_r = pd.read_csv('saudi24_race_laps.csv')
jed24_r
```

## Pick Data

```python
df1_jed24_r = jed24_r[['Time',
                       'Driver', 'DriverNumber', 
                       'LapNumber','Stint', 'LapTime', 'TrackStatus',
                       'PitInTime', 'PitOutTime',
                       'Sector1Time', 'Sector2Time', 'Sector3Time',
                       'SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST',
                       'Compound', 'TyreLife', 'FreshTyre'
                      ]]
df1_jed24_r   
```

## Clean Pit Time

```python
df1_jed24_r['PitTime'] = df1_jed24_r.apply(
    lambda x: x['PitInTime'] if pd.notnull(x['PitInTime']) else x['PitOutTime'], axis=1
)
df1_jed24_r
```

```python
df1_jed24_r.to_csv('df_jed24_r.csv', index=False)
```

```python
df2_jed24_r = df1_jed24_r.drop(columns=['PitInTime', 'PitOutTime'])
df2_jed24_r
```

```python
df2_jed24_r.info()
```

```python
df2_jed24_r.to_csv('df_jed24_r.csv', index=False)
```

## Convert time data

```python
# Define a function to convert timedelta strings to total seconds
def timedelta_to_seconds(timedelta_string):
    if pd.isnull(timedelta_string):
        return None
    td = pd.Timedelta(timedelta_string)
    return td.total_seconds()

# Convert time columns to seconds
time_columns = ['Time', 'LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time', 'PitTime']
for col in time_columns:
    df2_jed24_r[col] = df2_jed24_r[col].apply(timedelta_to_seconds)

df2_jed24_r
```

```python
df2_jed24_r.to_csv('df_jed24_r.csv', index=False)
```
