# Init

```python
import fastf1 as ff1
import pandas as pd

# Enable the cache
ff1.Cache.enable_cache('cache')
```

# Bahrain

## Bahrain'23

### Quali

```python
# Load the 2023 Bahrain race data
bahrain23_quali = ff1.get_session(2023, 'Bahrain', 'Q')

# Load the lap data for all drivers with tqdm as the progress bar
bahrain23_quali.load(telemetry=True)
bahrain23_quali.results
```

```python
bahrain23_quali.laps
```

```python
bahrain23_quali.laps.to_csv('bahrain23_quali_laps.csv', index=False)
```

### Race

```python
# Load the 2023 Bahrain race data
bahrain23_race = ff1.get_session(2023, 'Bahrain', 'R')

# Load the lap data for all drivers with tqdm as the progress bar
bahrain23_race.load(telemetry=True)
bahrain23_race.results
```

```python
bahrain23_race.laps.to_csv('bahrain23_race_laps.csv', index=False)
```
