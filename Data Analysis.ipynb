```python
plt.style.available
```

```python
# Create a figure and axis with a black background
fig, ax = plt.subplots(figsize=(14, 8))
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Loop through each team color in the dataframe
for color in jed24_q['TeamColor'].unique():
    # Filter the dataframe for each team color
    team_df = jed24_q[jed24_q['TeamColor'] == color]

    # Assuming there are two drivers per team
    drivers = team_df['Driver'].unique()

    for i, driver in enumerate(drivers):
        # Filter for each driver
        driver_df = team_df[team_df['Driver'] == driver]

        # Assign line style based on driver index
        line_style = '-' if i == 0 else '--'

        # Plot lap times for each driver using team color
        ax.plot(driver_df['LapNumber'], driver_df['LapTime'],
                line_style, label=f'{driver}', color=f'#{color}')

# Add a legend with white text
ax.legend(title='Driver', facecolor='black', edgecolor='white', labelcolor='white')

# Set the title and labels with white text
ax.set_title('Lap Times by Driver and Team Color', color='white')
ax.set_xlabel('Lap Number', color='white')
ax.set_ylabel('Lap Time (seconds)', color='white')

# Change the color of the axis ticks to white
ax.tick_params(colors='white')

# Show the plot
plt.show()
```

```python
# Get unique team names and set up the number of subplots in a 5x2 grid
team_names = jed24_q['TeamName'].unique()
n_rows = 5
n_cols = 2
fig, axs = plt.subplots(n_rows, n_cols, figsize=(20, 8 * n_rows), facecolor='black', squeeze=False)

# Flatten the axs array for easy iteration
axs = axs.flatten()

# Loop through each team name and create a subplot for each team
for i, team_name in enumerate(team_names):
    team_df = jed24_q[jed24_q['TeamName'] == team_name]
    team_color = team_df['TeamColor'].iloc[0]  # Assuming all rows for the team have the same color
    drivers = team_df['Driver'].unique()

    for j, driver in enumerate(drivers):
        driver_df = team_df[team_df['Driver'] == driver]
        line_style = '-' if j == 0 else '--'  # Solid line for first driver, dashed for the second
        axs[i].plot(driver_df['LapNumber'], driver_df['LapTime'], line_style, label=driver, color=f'#{team_color}')

    axs[i].set_title(f'Team: {team_name}', color='white')
    axs[i].set_facecolor('black')
    axs[i].tick_params(colors='white', which='both')
    axs[i].legend(title='Driver', facecolor='black', edgecolor='white', labelcolor='white')
    
    # Set grid to be dotted and a bit transparent
    axs[i].grid(visible=True, color='white', linestyle=':', linewidth=0.5, alpha=0.5)

# Set common labels and adjust layout for each subplot
for ax in axs:
    ax.set_xlabel('Lap Number', color='white')
    ax.set_ylabel('Lap Time (seconds)', color='white')

# Hide any unused subplots if there are fewer than 10 teams
for i in range(len(team_names), n_rows * n_cols):
    axs[i].axis('off')  # Turn off unused subplots

plt.tight_layout()
plt.show()
```

```python
# Calculate the fastest sector time for each driver
fastest_sector = df_pick.groupby('Driver')['Sector1Time'].min()

# Sort the results from fastest to slowest
sorted_fastest_sector = fastest_sector.sort_values()

# Get the team color for each driver
team_colors = df_pick.groupby('Driver')['TeamColor'].first()

# Create a horizontal bar plot with a black background
plt.figure(figsize=(10, 8), facecolor='black')
ax = plt.gca()  # Get current axis
ax.set_facecolor('black')

# Plot each driver's fastest sector time with their team color
for driver, time in sorted_fastest_sector.items():
    plt.barh(driver, time, color=f'#{team_colors[driver]}', zorder=3)

# Set scale to focus on smaller differences by setting x limits manually
# Adjust the x-axis range to have a smaller gap based on the minimum and maximum times
x_min = sorted_fastest_sector.min() - (sorted_fastest_sector.max() - sorted_fastest_sector.min()) * 0.05
x_max = sorted_fastest_sector.max() + (sorted_fastest_sector.max() - sorted_fastest_sector.min()) * 0.05
plt.xlim(x_min, x_max)

# Set the grid to be dotted and semi-transparent, behind the bars
plt.grid(True, which='both', color='white', linestyle='--', linewidth=0.5, alpha=0.5, zorder=2)

# Invert y-axis to have the fastest driver at the top, set axis labels, and title
ax.invert_yaxis()
plt.xlabel('Time (seconds)', color='white')
plt.title('Fastest Sector 1 Times by Driver', color='white')

# Change the color of all axis elements to white
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

# Show the plot
plt.show()
```

```python
# Sort the DataFrame to ensure that the data is in the right order.
df_sorted = df4_pick.sort_values(by=['Driver', 'LapNumber'])

# Shift the 'PitTime' column up by one row within each group of 'Driver'.
# This will align the pit exit 'PitTime' of each driver to the corresponding pit entry 'PitTime' row.
df_sorted['PitExitTime'] = df_sorted.groupby('Driver')['PitTime'].shift(-1)

# Calculate the pit duration time.
# This subtracts the pit entry 'PitTime' from the pit exit 'PitTime'.
df_sorted['PitDuration'] = df_sorted['PitExitTime'] - df_sorted['PitTime']

# Filter out the rows where the calculation would not make sense, such as the last lap of each driver.
df_sorted = df_sorted.dropna(subset=['PitDuration'])
df_sorted = df_sorted[df_sorted['PitDuration']<1000]
df_sorted
```

```python
# Calculate the mean Pit Duration for each driver
mean_pit_duration = df_sorted['PitDuration'].mean()

# Display the mean pit duration for each driver
mean_pit_duration
```
